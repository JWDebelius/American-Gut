# Power Analysis

While null hypothesis statistical testing demonstrates if something has an effect, effect sizes allow the estimation of how important that effect is. There have been many calls in medical literature for the inclusion of effect size along with statistical p values, as they increase the power of comparison [<a href="#17944619">1</a>,<a href="#23997866">2</a>]. We can leverage effect sizes within microbiome research in multiple ways. Understanding the effect size may help us rank which factors affect the microbiome the most. This, in turn, could help identify targets for intervention. Effect size can also help us design better studies. In the context of statistical power, effect size can help us estimate how many samples we might need to be reasonably confident that our data is true, given some margin of error.

The complex relation between humans and their microbes, especially the microbes in their guts, has been recognized in many areas of human health. Our microbes are known to change during the course of our lives, although most of this change occurs within the first three years [<a href="#20668239">3</a>-<a href="#22797518">6</a>]. Long term dietary patterns also have a large influence on our gut microbiome, although certain extreme dietary changes can force a switch [<a href="#21596990">7</a>-<a href="#24336217">9</a>]. The gut microbiome can also be re-shaped by antibiotic use [<a href="#20736229">10</a>-<a href="#25271726">13</a>]. It’s also known that the microbiome is altered in diseases. Inflammatory Bowel Disease (IBD) and Obesity are well studied examples in which the disease state is associated with a dysbiosis, or altered community structure [<a href="#25307765">14</a>-<a href="#24009397">25</a>]. Seasonal changes in the microbiome have also been previously reported [<a href="#24618913">26</a>]. We will demonstrate that in a healthy population of adults, age, alcohol consumption, exercise frequency and sleep duration also have impacts on the microbiome.

The list of factors which interact with the microbiome is based on the results of significance testing, and does not consider effect size. But, effect size is still an important consideration in microbiome research for both experimental design and possible interventions. Using the American Gut data, we can demonstrate that antibiotic use in the past month significantly decreases alpha diversity, compared with people who have not used antibiotics in the past year (p < 0.01). We can also show that people who drink alcohol regularly (three or more times per week) have higher alpha diversity than those who do not consume alcohol (p < 0.01).

If these results were published, one can imagine a situation in which popular media might advise the regular consumption of alcohol use following a dose of antibiotics to help repopulate the microbiome. If we ignore the number of logical fallacies associated with this practice, and instead assume that regular alcohol consumption is able to reseed the gut microbiome in a healthy way, the question then becomes if this will be an effective treatment. (Of course, this represents a potential example, and no real data beyond the significance exists to support this mechanistic hypothesis.) If the effect size of alcohol consumption is on the same order or greater than the effect of antibiotic use, than regular alcohol consumption following antibiotic use might be a good treatment strategy. On the other hand, if alcohol consumption has less effect than antibiotic use, the regular consumption of alcohol might not be worth the expense.

The American Gut data set present a rare opportunity of effect size prediction within the microbiome field. The large size and high degree of heterogeneity among participants and amount of survey metadata collected from each participant allows the examination of effect size for factors that are either currently uncharacterized or under characterized in microbiome research. The accompanying challenge is that effect size calculations are difficult for the type of data collected in microbiome research. Most traditional effect size metrics for comparing samples between groups, such as Cohen’s D, make assumptions about the normality of the data being studied, while microbiome data is not normal. As a result, we will leverage a method of empirically estimating statistical power.

### Statistical Hypotheses and Error: A "toy" example
Using statistics, we can test hypotheses. For example, we might want to test whether or not toy bricks costs different amounts depending on which model they come from. Perhaps we’re trying to build the most epic toy brick set to represent places where the microbiome is studied. We’re trying to do this in the cheapest way possible. We’ve determined that there are two relatively inexpensive sets which contain a specific brick we want: a chemistry laboratory and a komodo dragon enclosure. However, we’d also like to repurpose the rest of the bricks in the sets. So, we’d like to consider the cost per brick when we purchase the sets.

So, we’re going to test the hypothesis that the cost per brick in the lab playset is different from the cost per brick in the komodo dragon playset. We can write two possible outcomes for our statistical test, and call them hypotheses:
<center><strong>H<sub>0</sub></strong>: The model type is not related to the cost per brick.</center>
$$cost_{lab} = cost_{dragon}$$
<center><strong>H<sub>1</sub></strong>: The model type is related to the cost per brick.</center>
$$cost_{lab} \neq cost_{dragon}$$

In an ideal world, we would be able to collect information about every set sold in the country, which statisticians would call the population. But, presumably, this is a side project, and we don’t have enough time or energy to be able to collect information about every single toy brick set of a particular model sold at every single store in the country. 

Rather than collect information about every single set in the country, let’s assume that we were able to collect price information for the three toys a week at three stores across the country during the calendar year, 2013. Let’s call each time someone checked the price an “observation”. The total set of observations of each model is a “sample”. For each observation, we have the location, calendar date, and whether or not there was an advertised sale. Using population based data from the US census bureau, we can extrapolate more information about the community. Since we have a small number of observations at each point, we can trade resolution for sample number and pool the data about our models into assigned categories like **SEASON**, **DATE**, **REGION**, **COMMUNITY_TYPE**, **STORE** and **SALE**.

When we go to test our hypothesis about the model influencing the cost of bricks, there are four possible outcomes of our experiment:

Our statistical test can make a mistake in two ways. We can reject H<sub>0</sub> when it is, in fact true, leading to a False positive. In our toy example, this would be saying that the bricks within the two sets cost different amounts, when in fact, they have the same cost. In a medical context, this might mean saying that someone has cancer, when in fact, they are healthy. Statisticians try to limit the risk associated with False Positives by setting their critical value, called α, which is the acceptable probability that a False positive has happened. Statistical p values are compared to the value of α to determine if the risk of a false positive is acceptable.

We can also make a mistake and fail to reject H<sub>0</sub> when it is True. In the toy example, this would mean we say that the cost of the bricks in the two set are the same, when they are different. In the cancer example, it would mean saying a patient is cancer free when they have the disease. This is also dangerous, but more difficult to test. The statistical power of an experiment is 1 - (probability of a false negative). Effect size is related to the rates of both false positives and false negatives, along with the number of samples present.

### A More Formal Mathematical Defination for Power Analysis
Traditionally, statistical power can only be calculated for certain types of data and certain statistical tests. Most effect size measures make assumptions about the distribution of the data (for example, Cohen’s F is the power of an ANOVA, which assumes a normal distribution)[<a href="#Cohen">27</a>,<a href="#Zar">28</a>]. Unfortunately, most microbiome data violates one or more traditional assumptions. For example, many bacterial taxa are sparsely represented and therefore do not follow a normal distribution [<a href="#23284876">29</a>]. One approach to overcome problems associated with non-normal distributions is to use permutations [<a href="#Bondini">30</a>]. We have applied a permutative approach to empirically estimate power for microbiome data.

#### Defination
Let's assume there exists a set of populations, $\{K_{1}, K_{2}, ... K_{k}\}$, which have some parameter, $\chi$ where $\chi_{1} \neq \chi_{2} \neq \chi{k}$. For each of these populations, let a random samples be drawn, $\{S_{1}, S_{2}, ... S_{k}\}$ of sizes $\{N_{1}, N_{2}, ... N_{k}\}$ and let each sample, $S$, be characterized by some parameter, $X$ where $\chi_{i} \cong X_{i}$.
For our set of samples, we can test the hypotheses:
$$\textrm{H}_{\textrm{0}}\textrm{:  } \chi_{1} = \chi_{2} = ... = \chi_{k}$$
$$\textrm{H}_{\textrm{1}}\textrm{:  } \chi_{1} \neq \chi_{2} \neq ... \neq \chi_{k}$$
using some statistical test, $\mathbf{F}$. For the set of samples, $\mathbf{F}(S) = \alpha$, where for some critical value, $p_{crit}$, $\alpha < p_{crit}$, indicating we should reject H<sub>0</sub>.

If $N_{i}$ is suffeciently large, we can randomly subsample $S_{i}$ to generate $s_{i,j}$, a subsample of size $n_{j}$ where $n_{j} < N{i}$. Each subsample will have a parameter, $x_{i,j}$ where $x_{i,j} \cong X_{i}$, which by the transitive property, indicates that $x_{i,j} \cong \chi_{i}$. 

We can apply our statical test, $\mathbf{F}$ to $m$ rounds of random subsampling at each sample size to generate a set of p values, $\{p_{1}, p_{2}, ..., p_{m}\}$.<br>
We assume that our samples are random and representative, and we know that we should reject H<sub>0</sub> because $\alpha < p_{crit}$ than for each of the representative subsets, $p_{j} < p_{crit}$. If $p_{j} \geq p_{crit}$, we can assume a type II error has occured.

We can define the emperical probability of comitting a type II error, $\beta$ as
$$\beta = \frac{\left | p_{j} < p_{crit}\right |}{m} \textrm{  for  } \{j | j \textrm{ } \epsilon \textrm{ }\mathbb{N} \textrm{ and } 1 \leq j \leq m\}$$
where $\left | a \right |$ is the number of elements in set <em>a</em>.

Since the statistical power, $P$, is defined as 
$$P = 1 - \beta$$
we can say that
$$P = 1 - \left ( \frac{\left | p_{j} < p_{crit}\right |}{m} \right )$$
$$P = \left ( \frac{\left | p_{j} > p_{crit}\right |}{m} \right )$$

If we repeat the subsampling process $g$ times, the central limit theorem applied to weak convegence says the mean of the power estimate, $\overline{P} \rightarrow P$.

### Proof of Principle
The method of power analysis we’re using here is a relatively new method. We wanted to demonstrate the empirical method give similar results to traditional power analysis techniques. We applied a case II student’s t test, which compares the hypotheses,
<center><strong>H<sub>0</sub></strong>: The means are equal</center>
$$\mu_{1} = \mu_{2}$$
<center><strong>H<sub>1</sub></strong>: The means are not equal.</center>
$$\mu_{1} \neq \mu_{2}$$

where $\mu$ is the mean of a normal distribution. 

The power of this test is given by
$$PWR(\mu_{1}, \mu_{2}, \sigma_{1}^{2}, \sigma_{2}^{2}, n_{1}, n_{2}, \alpha) = P \left (Z > -z_{1 - \alpha/2} + \frac{\sqrt{(\mu_{1} - \mu_{2})^{2}}}{\sigma_{1}^{2}/n_{1} + \sigma_{2}^{2}/n_{2}} \right )$$
$$PWR(\mu_{1}, \mu_{2}, \sigma_{1}^{2}, \sigma_{2}^{2}, n_{1}, n_{2}, \alpha) = pnorm \left ( -z_{1 - \alpha/2} + \frac{\sqrt{(\mu_{1} - \mu_{2})^{2}}}{\sigma_{1}^{2}/n_{1} + \sigma_{2}^{2}/n_{2}}, 0, 1 \right )$$

where $pnorm(x, 0, 1)$ is the probability density function of $x$ within a normal distribution with mean 0 and standard deviation 1. 

To simplify the problem, we simulated two normal distributions with the same standard deviation. The standard deviation was 1.5 times larger than the difference between the means. We also drew the same number of observations from each distribution, which allowed us to simplify the power equation:
$$PWR(\bar{x}_{1}, \bar{x}_{2}, s_{1}, s_{2}, n, \alpha) = pnorm \left(-z_{1 - \alpha/2} + \sqrt{\frac{n(\bar{x}_{1} + \bar{x}_{2})^{2}}{s_{1}^{2} + s_{2}^{2}}} \right )$$

We drew samples of 25, 50, 100, and 1000 observations from each of the two distributions, and compared the results with a t-test. The rejection region is highlighted on each sample with the dotted black line and gray shading in the histograms. The samples were used to generate statistcal power in three ways. First, we calculated power using the traditional method (solid line). Emperical power was determined by subsampling (blue circles). The empercial power was then used to extrapolate a curve (dashed blue line).

![sample power comparison](https://github.com/JWDebelius/American-Gut/blob/rounds_1_14_notebook/ipynb/images/power_subsample.png?raw=true)

There is fairly good agreement between the traditional and elucidated power curves, especially at higher levels of power. By 50% power, the two curves are within a 10% of each other (the extrapolated curve overestimates power at lower sample sizes). Some of the difference in sizes may be associated with  This suggests that for a first-pass methodology for estimating sample size, calculation of empirical power curves, and extrapolating beyond sampling is a reliable method and represents an improvement on currently available methods.

### Caveots and Considerations
The power analysis we’re performing here assume the following things about our data and statistical tests:

1. The samples are a representative, random sample of the underlying population.

2. Samples are sufficiently large to allow random subsampling at a depth that faciliates power analysis. For some statistical tests, a subsample of less than 5 observations per group may not be appropriate.

3. There is a significant difference between the samples, or reason to believe there is a significant difference between the populations.

4. Samples must satisfy any requirements of the statistical test, $F$.

### Alpha Diversity Parameters

We’re also going to define a function for alpha diversity power analysis. We will test our alpha diversity using a <a href="http://en.wikipedia.org/wiki/Kruskal–Wallis_one-way_analysis_of_variance">Kruskal-Wallis test</a>. A Kruskal-Wallis test is non-parametric and does not require normally distributed data. Our function will take the alpha diversity value for each group, and then compare them using the test.

```python
# Defines the alpha diversity test function
def test_alpha_kruskal(ids, meta, a_div_metric):
    """Calculates difference in alpha diverisity for a category"""
    # Gets the alpha diveristy values at the location given
    alpha = [meta.loc[id_, a_div_metric] for id_ in ids]
    # Calculates the p value by comparing categories
    return scipy.stats.kruskal(*alpha)[1]
```

We test for differences in beat diversity using the scikit-bio function, [permanova](http://scikit-bio.org/docs/latest/generated/generated/skbio.stats.distance.permanova.html#skbio.stats.distance.permanova). We’ll combine all the sample ids, and then use a map and distance matrix with only these observations.

```python
def test_beta_permanova(ids, meta, dm, cat, num_iter=499):
    """Tests difference in beta diversity for a category and distance matrix"""
    # Gets the map and distance matrix subset
    all_ids = np.hstack(ids)
    # Calculates the permanova
    perma_results = skbio.stats.distance.permanova(dm.filter(all_ids), 
                                                   meta.loc[all_ids],
                                                   cat,
                                                   num_iter)
    # Returns the p value
    return perma_results['p-value']
```

### General Parameters for Power Analysis
The empirical power calculation method in [scikit-bio](http://scikit-bio.org/docs/0.2.2/generated/skbio.stats.power.html) is based on iteration. It works by matching pairs of observations based on a set of metadata categories. A number of observations (set using <strong><code>min_counts</code></strong>, <strong><code>max_counts</code></strong>, and <strong><code>counts_interval</code></strong>) are then drawn at random from this set of matched data, and a statistical test is performed. This is repeated a number of times (given by <strong><code>num_iter</code></strong>), and the fraction of samples below a critical value (<strong><code>p_crit</code></strong>) is calculated as the power at that sample depth. This is repeated several times to determine the empirical power with some degree of confidence.

```python
import time
class timewith():
    def __init__(self, name=''):
        self.name = name
        self.start = time.time()

    @property
    def elapsed(self):
        return time.time() - self.start

    def checkpoint(self, name=''):
        print '{timer} {checkpoint} took {elapsed} seconds'.format(
            timer=self.name,
            checkpoint=name,
            elapsed=self.elapsed,
        ).strip()

    def __enter__(self):
        return self

    def __exit__(self, type, value, traceback):
        self.checkpoint('finished')
        pass
```

## Discussion
From our empirical effect size estimations, we can being to rank effects on the microbiome. As demonstrated by the Human Microbiome Project (HMP), body site has the largest effect on the human microbiome [<a href="#22699609">31</a>].

We find that an IBD diagnosis has one of the largest impacts on the gut microbiome. While dysbiosis in IBD is likely a complex relationship between the host genetics, host lifestyle and the microbiome, the disease is characterized by a profoundly altered microbiome and intestinal inflammation [<a href="#25307765">14</a>-<a href="#25567118">20</a>]. As with IBD, the microbiome is considered strong predictive of obesity [<a href="#19043404">21</a>-<a href="#24009397">25</a>]. However, we find the effect of obesity on the microbiome is smaller than lifestyle factors like sleep and exercise for both alpha and beta diversity. This finding is corroborated, to some degree, by a recent meta analysis which found IBD-related microbial signatures to be consistent across studies while obesity-related dysbiosis was more strong influence by technical effects [<a href="#25307765">14</a>].

We might have predicted the difference in magnitude of these effects based gut pathology. IBD has been recognized as a disease of the intestines and a diagnosis is based on intestinal pathology [<a href="#19653289">35</a>,<a href="#21530745">36</a>]. In contrast, a diagnosis of obesity cannot be made by looking at the gut alone. It’s likely the complex interactions of IBD-specific bacteria and bacterial strains along with host inflammation shapes the community structure in a unique way. It seems less likely we will see such a strong disease-associated effect in conditions that do not traditionally have gut involvement. It might be more reasonable to expect they will have an effect size closer to obesity.

We also find that antibiotic use has a large effect on both alpha and beta diversity. This has been previously recognized [<a href="#20847294">11</a>-<a href="#25271726">13</a>]. Intuitively, it’s also not entirely unexpected; the goal of antibiotics are to kill bacteria. Therefore, a systemic dose of antibiotics will likely come with off-target effects.

Among lifestyle factors, we see the one dietary variable, the number of types of plants consumed in a week, has the largest influence. This effect is larger than the effect of recent antibiotic use. Long term dietary patterns are known to have a strong effect on the microbiome [<a href="#22797518">6</a>-<a href="#21885731">8</a>]. In conjunction with this finding, we also observe that daily alcohol consumption has a smaller effect on the microbiome than plant consumption, or age, but a larger effect than obesity. We also found that sleep duration has a moderate effect. Individuals who sleep less six hours or less per night have less diversity and shifted microbiomes than those who sleep more than eight hours per night. This effect is larger than the effect of obesity. Circadian dysfunction has been linked to obesity, although the mechanism frequently cited is a disruption of mammalian core clock genes [<a href="#23434278">37</a>]. Exercise is known to have an effect on the microbiome, and the effect of daily exercise, compared to those who exercise less than once a week, is noticeable, but small in the scheme of effects investigated here [<a href="#25021423">38</a>,<a href="#24670791">39</a>]. The effect of daily exercise on beta diversity is of the same order of magnitude as the effect of obesity, while the effect on alpha diversity is much larger. 

We also find strong effects associated with temporal variation. Changes in the microbiome at the extremes of the human lifespan - from early development to the effects of aging on the microbiome have been well studied [<a href="#20668239">3</a>-<a href="#22797518">6</a>]. The most extreme changes in the microbiome occur during the first three years of life, and involves changes in body site composition [<a href="#20668239">3</a>,<a href="#22699611">4</a>,<a href="#20566857">40</a>]. We see a strong effect of age in adults: age has almost as large an effect on beta diversity as antibiotic use, although the effect on alpha diversity is much smaller. 

We also see a seasonal effect, which was previously observed in a population of 60 matched samples [<a href="#24618913">26</a>]. The effect of collection season on the microbiome is on the same order as the effect we see for sleep duration, and larger than the effect seen for obesity. This may encourage a shift in experiment design. At a minimum, it suggests that tracking collection date in human microbiome studies may be important. 
Our power analysis allows us to rank the effect sizes of disease, lifestyle, and temporal factors on the human microbiome. We find that effect sizes are generally larger in beta diversity than alpha diversity, suggesting that most changes associated with the extreme states are caused by changes in the taxa present in communities between the two states. A more concerning aspect to the power analysis, however, suggests that many cross-sectional microbiome studies currently in the literature are underpowered. We hope the effect size predictions here can serve as a guide for future studies, and encourage appropriate selection of populations.
