```python
questions = ['IBD', 'SEX', 'ANTIBIOTIC_HISTORY', 'COUNTRY', 'RACE', 'BOWEL_MOVEMENT_QUALITY',
             'BOWEL_MOVEMENT_FREQUENCY', 'AGE_CORRECTED']
```

```python
fecal_data.map_.AGE_CORRECTED.min()
```

```python
for name in questions[:-1]:
    question = agdic.ag_dictionary(name)
    if name == 'COUNTRY':
        question.order = ['USA', 'Australia', 'Canada', 'United Kingdom']
    elif name == 'IBD':
        question.order = ['No', 'Yes']
    fecal_data.clean_group(question)
    if name == 'AGE_CORRECTED':
        continue
    question.label_order(fecal_data.map_)
map_ = fecal_data.map_.loc[fecal_data.map_['AGE_CORRECTED'] > 19]
```

```python
actual = set(ibd).intersection(map_.index)
map_.loc[actual, 'IBD'] = '(0) Yes'
```

```python
vios_skip = {'VIOSCREEN_FINISHED',
             'VIOSCREEN_DATABASE',
             'VIOSCREEN_MULTIVITAMIN',
             'VIOSCREEN_QUESTIONNAIRE',
             'VIOSCREEN_GENDER',
             'VIOSCREEN_SCFV',
             'VIOSCREEN_PROTOCOL',
             'VIOSCREEN_CALCIUM_FREQ',
             'VIOSCREEN_CALCIUM',
             'VIOSCREEN_STARTED',
             'VIOSCREEN_NUTRIENT_RECOMMENDATION',
             'VIOSCREEN_PROCDATE',
             'VIOSCREEN_DOB',
             'VIOSCREEN_SCF',
             'VIOSCREEN_MULTIVITAMIN_FREQ'}
vios_cols = [x for x in map_.columns if 'VIOSCREEN' in x and not x in vios_skip]
for col in vios_cols:
    map_[col] = map_[col].astype(float)
```
